# Drift velocity and electron diffusion models

The electron cloud drift and diffusion processes are mainly dependent on the local electric field. In a first approximation, the local electric field is established as uniform and denoted only by *drift field*. The dirft field.dependent models implemented in `xenodiffusionscope` are the models hardcoded in **NEST v2.0.0**.

```python
import matplotlib.pyplot as plt
import numpy as np

import xenodiffusionscope as xds
```

```python
def plot_drift_speed(tpc, drift_fields, ax):
    ax.plot(drift_fields, tpc.model_velocity(drift_fields), c = 'k')
    ax.set_xlabel('Drift field [V/cm]')
    ax.set_ylabel('Drift speed [mm/us]')
    ax.set_xscale('log')
    ax.grid()
    return ax

def plot_diffusion_long(tpc, drift_fields, ax):
    ax.plot(drift_fields, tpc.model_diffusion_longitudinal(drift_fields), c = 'k')
    ax.set_xlabel('Drift field [V/cm]')
    ax.set_ylabel('Longitudinal diffusion [mm$^2$/us]')
    ax.set_xscale('log')
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
    ax.grid()
    return ax

def plot_diffusion_transv(tpc, drift_fields, ax):
    ax.plot(drift_fields, tpc.model_diffusion_transversal(drift_fields), c = 'k')
    ax.set_xlabel('Drift field [V/cm]')
    ax.set_ylabel('Transversal diffusion [mm$^2$/us]')
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
    ax.set_xscale('log')
    ax.grid()
    return ax
```

```python
# Create TPC object to access class methods
Xenoscope = xds.TPC

# Define drift field values to plot
drift_fields = np.logspace(1,3,100)
```

```python
fig, axs = plt.subplots(1,3, figsize = (15,5), dpi = 60)
axs[0] = plot_drift_speed(Xenoscope, drift_fields, axs[0])
axs[1] = plot_diffusion_long(Xenoscope, drift_fields, axs[1])
axs[2] = plot_diffusion_transv(Xenoscope, drift_fields, axs[2])
fig.tight_layout()
plt.show()
```
